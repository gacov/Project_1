```python
# Load file
datafile20 = "Resources/2020.csv"
datafile19 = "Resources/2019.csv"
datafile18 = "Resources/2018.csv"
datafile17 = "Resources/2017.csv"
datafile16 = "Resources/2016.csv"
datafile15 = "Resources/2015.csv"

# Read 202 File and store into Pandas data frame
yr20df = pd.read_csv(datafile20)
yr19df = pd.read_csv(datafile19)
yr18df = pd.read_csv(datafile18)

yr17df = pd.read_csv(datafile17)
yr16df = pd.read_csv(datafile16)
yr15df = pd.read_csv(datafile15)
```

```python
cleanedyr20df = yr20df[["Country name","Ladder score","Explained by: Log GDP per capita","Explained by: Social support","Explained by: Healthy life expectancy","Explained by: Freedom to make life choices","Explained by: Generosity","Explained by: Perceptions of corruption"]]
cleanedyr20df = cleanedyr20df.rename(columns={"Country name":"Country",
                      "Ladder score":"Score 20",
                      "Explained by: Log GDP per capita":"GDP per capita 20",
                      "Explained by: Social support":"Social Support 20",
                    "Explained by: Healthy life expectancy":"Healthy life expectancy 20",
                    "Explained by: Freedom to make life choices":"Freedom to make life choices 20",
                    "Explained by: Generosity":"Generosity 20",
                    "Explained by: Perceptions of corruption":"Perceptions of corruption 20"
                     })
cleanedyr20df.shape
```

```python
yr19df = yr19df.rename(columns={"Country or region":"Country"})
```

```python
cleanedyr20and19df = pd.merge(cleanedyr20df,yr19df,how='left',on ='Country')
cleanedyr20and19df = cleanedyr20and19df.rename(columns={
                      "Score":"Score 19",
                      "GDP per capita":"GDP per capita 19",
                      "Social support":"Social Support 19",
                    "Healthy life expectancy":"Healthy life expectancy 19",
                    "Freedom to make life choices":"Freedom to make life choices 19",
                    "Generosity":"Generosity 19",
                    "Perceptions of corruption":"Perceptions of corruption 19"
                     })
cleanedyr20and19df
```

```python
yr18df = yr18df.rename(columns={"Country or region":"Country"})
```

```python
cleaned3yrsdf = pd.merge(cleanedyr20and19df,yr18df,how='left',on ='Country')
cleaned3yrsdf
```

```python
cleaned3yrsdf = cleaned3yrsdf.rename(columns={
                      "Score":"Score 18",
                      "GDP per capita":"GDP per capita 18",
                      "Social support":"Social Support 18",
                    "Healthy life expectancy":"Healthy life expectancy 18",
                    "Freedom to make life choices":"Freedom to make life choices 18",
                    "Generosity":"Generosity 18",
                    "Perceptions of corruption":"Perceptions of corruption 18"
                     })
del cleaned3yrsdf["Overall rank_x"]
del cleaned3yrsdf["Overall rank_y"]
```

```python
cleanedyr17df = yr17df[["Country","Happiness.Score","Economy..GDP.per.Capita.","Family","Health..Life.Expectancy.",
                        "Freedom","Generosity","Trust..Government.Corruption."]]
cleanedyr17df = cleanedyr17df.rename(columns={"Happiness.Score":"Score 17",
                      "Economy..GDP.per.Capita.":"GDP per capita 17",
                      "Family":"Social Support 17",
                    "Health..Life.Expectancy.":"Healthy life expectancy 17",
                    "Freedom":"Freedom to make life choices 17",
                    "Generosity":"Generosity 17",
                    "Trust..Government.Corruption.":"Perceptions of corruption 17"
                     })
cleanedyr17df
```

```python
cleanedyr16df = yr16df[["Country","Happiness Score","Economy (GDP per Capita)","Family","Health (Life Expectancy)",
                        "Freedom","Generosity","Trust (Government Corruption)"]]
cleanedyr16df = cleanedyr16df.rename(columns={"Happiness Score":"Score 16",
                      "Economy (GDP per Capita)":"GDP per capita 16",
                      "Family":"Social Support 16",
                    "Health (Life Expectancy)":"Healthy life expectancy 16",
                    "Freedom":"Freedom to make life choices 16",
                    "Generosity":"Generosity 16",
                    "Trust (Government Corruption)":"Perceptions of corruption 16"
                     })
cleanedyr16df
```

```python
cleanedyr15df = yr15df[["Country","Happiness Score","Economy (GDP per Capita)","Family","Health (Life Expectancy)",
                        "Freedom","Generosity","Trust (Government Corruption)"]]
cleanedyr15df = cleanedyr15df.rename(columns={"Happiness Score":"Score 15",
                      "Economy (GDP per Capita)":"GDP per capita 15",
                      "Family":"Social Support 15",
                    "Health (Life Expectancy)":"Healthy life expectancy 15",
                    "Freedom":"Freedom to make life choices 15",
                    "Generosity":"Generosity 15",
                    "Trust (Government Corruption)":"Perceptions of corruption 15"
                     })
cleanedyr15df
```
